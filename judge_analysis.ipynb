```python
narr, parr = [0]*14, [0]*14
correct_answer_traces = set([2, 4, 10, 12, 18, 20, 21, 22, 23, 26, 30, 33, 36, 41, 45, 47, 48, 50])
for i in range(1, 51):
    arr = []
    with open("llmjudge_output/llmjudge_output_"+str(i)+".txt") as f:
        for line in f:
            line = line.strip()
            if line.startswith('[') and line.endswith(']'):
                arr.append(int(line[1:-1]))
    if i in correct_answer_traces:
        for i in range(len(arr)):
            parr[i] += arr[i]
    else:
        for i in range(len(arr)):
            narr[i] += arr[i]

print(parr)
print(narr)
```

```python
import matplotlib.pyplot as plt
import numpy as np

def plot_fm_dist(vals, title):

    # Your data
    values = vals

    # Custom labels for each bar
    labels = [
        "1.1", "1.2", "1.3", "1.4", "1.5", "2.1", "2.2",
        "2.3", "2.4", "2.5", "2.6", "3.1", "3.2", "3.3"
    ]

    # Create bar chart
    plt.bar(labels, values)

    # Add labels and title
    plt.xlabel('Failure Modes')
    plt.ylabel('Occurences')
    plt.title(title)
    plt.ylim(0, 1.1)

    # Optional: Rotate labels if there are many
    plt.xticks(rotation=45)

    # Show the chart
    plt.show()
```

```python
parr = [1, 0, 4, 1, 1, 1, 0, 1, 1, 1, 1, 1, 1, 1]
narr = [0, 0, 1, 0, 0, 1, 0, 1, 0, 0, 0, 0, 1, 0]
nps = 27
nns = 1
```

```python
plot_fm_dist(np.array(parr)/nps, "Failure Mode Distribution for Successful Answers")
```

```python
plot_fm_dist(np.array(narr)/nns, "Failure Mode Distribution for Unsuccessful Answers")
```
