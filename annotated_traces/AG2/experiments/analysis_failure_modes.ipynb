```python
from pandas import read_csv
import matplotlib.pyplot as plt
```

```python
df = read_csv("../annotations.csv", index_col=0)
```

```python
df.loc[:, df.columns[3:]] = df.loc[:, df.columns[3:]].apply(lambda x: x=="yes")
```

```python
for c in df.columns[3:]:
    if df[c].sum() < 0.5:
        del df[c]
```

```python
true_counts = df.iloc[:, 3:].sum().sort_values(ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(true_counts.index, true_counts.values)
plt.xlabel('Failure Modes', fontsize=14)
plt.ylabel('Cases', fontsize=14)  
plt.title('Failure modes in AG2 - Scenario MathChat', fontsize=16)  
plt.xticks(rotation=45, ha='right', fontsize=12)
plt.yticks(fontsize=12)  
plt.tight_layout()
plt.show()
```
